```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Adjusting matplotlib settings for larger font sizes
plt.rcParams.update({'font.size': 16})  # Adjust font size globally
plt.rcParams.update({'axes.titlesize': 16})  # Adjust title size
plt.rcParams.update({'axes.labelsize': 16})  # Adjust axis label size

# Data for Patient 1
mean_z_data_p1 = {
    "Cluster 1": [0.857370, -0.290599, -0.619394],
    "Cluster 2": [-0.492951, -0.099646, 0.623732],
    "Cluster 3": [-0.278769, 1.001362, -0.708468]
}
std_dev_data_p1 = {
    "Cluster 1": [0.880254, 0.785725, 1.010229],
    "Cluster 2": [0.742859, 0.893724, 0.560461],
    "Cluster 3": [0.702561, 1.176288, 0.790522]
}

# Data for Patient 2
mean_z_data_p2 = {
    "Cluster 1": [0.420350, 0.363503, -1.005202],
    "Cluster 2": [-0.717625, 0.081829, 0.690189],
    "Cluster 3": [-0.164065, -0.945666, 0.753898]
}
std_dev_data_p2 = {
    "Cluster 1": [1.112853, 1.034159, 0.814553],
    "Cluster 2": [0.679897, 0.529563, 0.393786],
    "Cluster 3": [0.425418, 0.686298, 0.395590]
}

# Create DataFrames
index_labels = ["McGill Pain Questionnaire (total)", "Heart Rate", "Minutes Asleep"]
mean_z_df_p1 = pd.DataFrame(mean_z_data_p1, index=index_labels)
std_dev_df_p1 = pd.DataFrame(std_dev_data_p1, index=index_labels)
mean_z_df_p2 = pd.DataFrame(mean_z_data_p2, index=index_labels)
std_dev_df_p2 = pd.DataFrame(std_dev_data_p2, index=index_labels)

# Creating 2x2 plots with specified label visibility and color schemes
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Heatmap for mean z-score data for Patient 1
sns.heatmap(mean_z_df_p1, annot=True, cmap="RdBu_r", center=0, ax=axes[0, 0], vmin=-1.5, vmax=1.5)
axes[0, 0].set_title("Patient 1: Mean Z-Scores Heatmap")
axes[0, 0].set_yticklabels(index_labels, rotation=0)
axes[0, 0].set_xticklabels([])  # Hide cluster labels
axes[0, 0].set_xlabel("")  # Hide x-axis label

# Heatmap for standard deviation data for Patient 1
sns.heatmap(std_dev_df_p1, annot=True, cmap="gray_r", center=0, ax=axes[0, 1], vmin=-1.5, vmax=1.5)
axes[0, 1].set_title("Patient 1: Mean Standard Deviations Heatmap")
axes[0, 1].set_yticklabels([])
axes[0, 1].set_xticklabels([])
axes[0, 1].set_xlabel("")  # Hide x-axis label
axes[0, 1].set_ylabel("")  # Hide y-axis label

# Heatmap for mean z-score data for Patient 2
sns.heatmap(mean_z_df_p2, annot=True, cmap="RdBu_r", center=0, ax=axes[1, 0], vmin=-1.5, vmax=1.5)
axes[1, 0].set_title("Patient 2: Mean Z-Scores Heatmap")
axes[1, 0].set_yticklabels(index_labels, rotation=0)
axes[1, 0].set_xticklabels(mean_z_df_p2.columns, rotation=0)
axes[1, 0].set_xlabel("Clusters")  # Show cluster labels

# Heatmap for standard deviation data for Patient 2
sns.heatmap(std_dev_df_p2, annot=True, cmap="gray_r", center=0, ax=axes[1, 1], vmin=-1.5, vmax=1.5)
axes[1, 1].set_title("Patient 2: Mean Standard Deviations Heatmap")
axes[1, 1].set_yticklabels([])
axes[1, 1].set_xticklabels(std_dev_df_p2.columns, rotation=0)
axes[1, 1].set_xlabel("Clusters")  # Show cluster labels
axes[1, 1].set_ylabel("")  # Hide y-axis label

plt.tight_layout()
plt.savefig('pts_1_and_2_heatmap_plots.png', format='png', dpi=300)
plt.savefig('pts_1_and_2_heatmap_plots.svg', format='svg', dpi=300)

plt.show()
```
